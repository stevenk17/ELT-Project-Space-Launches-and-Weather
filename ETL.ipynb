```python
import pandas as pd 
from sqlalchemy import create_engine
```

 ### Store CSV into DataFrame

```python
#Import CSV File
csv_file="Resources/SpaceMissions.csv"

#Store CSV File in Dataframe
space_missions_df=pd.read_csv(csv_file)

#Read Dataframe
space_missions_df.head()
```

### Create new data  with selected column

```python
#Creating new dataframe with selected columns" Company, Launch Date, Launch Time, Launch Site, and Vehicle Type" 
new_space_missions_df = space_missions_df[["Company","Launch Date","Launch Time","Launch Site","Vehicle Type"]]

#Highlight any nans with the color red. 
new_space_missions_df.style.highlight_null(null_color='red')

```

```python
#Rename dataframe column from launch time to Launch Time Min:Sec
spacemissions_df=new_space_missions_df.rename(columns={"Launch Time":"Launch Time (Min:Sec)"})

#Remove nans with code below
spacemissions_df=spacemissions_df.dropna(how="any")

#Read dataframe as .style to show that nans were removed 
spacemissions_df.style
```

```python
#Change string in launch date  as datetime64[ns] to work
spacemissions_df['Launch Date'] = spacemissions_df['Launch Date'].astype('datetime64[ns]')

#Sort by date. 
spacemissions_df=spacemissions_df.sort_values(by='Launch Date')

#read as .style to show sorted columns
spacemissions_df.style
```
